```python
# Importer les bibliothèques nécessaires
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Charger les données de chiffres manuscrits
digits = load_digits()

# Diviser les données en ensembles de formation et de test
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Créer un classificateur KNN avec k=5
knn = KNeighborsClassifier(n_neighbors=5)

# Adapter le classificateur aux données de formation
knn.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = knn.predict(X_test)

# Calculer la précision des prédictions
accuracy = accuracy_score(y_test, y_pred)

# Afficher la précision des prédictions
print("Précision : {:.2f}%".format(accuracy * 100))
```

```python
import matplotlib.pyplot as plt

# Définir une plage de valeurs pour k
k_values = range(1, 31)

# Initialiser une liste vide pour stocker les précisions
accuracies = []

# Boucle sur les différentes valeurs de k
for k in k_values:
    # Créer un classificateur KNN avec k actuel
    knn = KNeighborsClassifier(n_neighbors=k)

    # Adapter le classificateur aux données de formation
    knn.fit(X_train, y_train)

    # Faire des prédictions sur l'ensemble de test
    y_pred = knn.predict(X_test)

    # Calculer la précision des prédictions
    accuracy = accuracy_score(y_test, y_pred)

    # Ajouter la précision à la liste des précisions
    accuracies.append(accuracy)

# Tracer le graphique de précision en fonction de k
plt.plot(k_values, accuracies)
plt.title('Précision en fonction de la valeur de k')
plt.xlabel('Valeur de k')
plt.ylabel('Précision')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
digits = load_digits()
X, y = digits.data, digits.target
```

```python
fig, ax = plt.subplots(2, 5, figsize=(10, 5))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='gray_r')
    axi.set(xticks=[], yticks=[], xlabel=digits.target[i])
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
